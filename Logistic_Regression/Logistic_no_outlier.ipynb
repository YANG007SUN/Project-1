```python
import pandas as pd
import numpy as np
import pathlib
from pprint import pprint
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import math

csvpath = "master_no_outlier.csv"
df = pd.read_csv(csvpath)
```

```python

df = df.drop(["yelp_rating_zscore", "lng_zscore", "name 2_zscore", "Walk Score_zscore", "Unemployment_zscore", "Household Income_zscore", "lat_zscore", "review_count_zscore", "Unnamed: 0", "Population_zscore", "Median Age_zscore", "Per Capita Income_zscore", "Poverty Count_zscore",  "Bike Score_zscore", "Midwest_zscore", "South_zscore", "West_zscore"], axis=1)
```

```python
df
```

```python
new_df = df
```

```python
new_df
```

```python
### Train Data
X= new_df.drop("name 2",axis=1)
y = new_df["name 2"]
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions  = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
pprint(classification_report(y_test,predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,predictions)
```

```python
from sklearn.utils import shuffle
y_test_shuf = shuffle(y_test)
```

```python
accuracy_score(y_test_shuf,predictions)
```
