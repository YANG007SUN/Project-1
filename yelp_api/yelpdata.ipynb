```python
import pandas as pd
import requests
import json
from config import api_key
from pprint import pprint
import time
```

```python
# import panda express dataset
px_data= pd.read_csv("../dataset/px_data.csv")
px_data
```

```python
# construct endpoint and header
endpoint_match = "https://api.yelp.com/v3/businesses/matches"
header ={"Authorization": "Bearer %s" % api_key}
params = {
    "name":"Panda Express",
    "country":"US",
    "limit":"1"
}
```

```python
# for each store, find out specific id
for index, row in px_data.iterrows():
    
    # parameters
    params["address1"] = row[5]
    params["city"] = row[6]
    params["state"] = row[7]
    params["latitude"] = row[3]
    params["longitude"] = row[4]
    
    try:
        # request from yelp
        response = requests.get(endpoint_match, params=params, headers = header).json()
        px_data.loc[index, "business name"] = response["businesses"][0]["name"]
        px_data.loc[index, "id"] = response["businesses"][0]["id"]
    
    except:
        
        # due to internal error message, use another try except block to re-run requests
        try:
            response = requests.get(endpoint_match, params=params, headers = header).json()
            px_data.loc[index, "business name"] = response["businesses"][0]["name"]
            px_data.loc[index, "id"] = response["businesses"][0]["id"]
        except:
            print("-------------------------------------------------------------")
            print(f"Store {row[0]} in City: {row[6]}, State: {row[7]} did not find.")
```

```python
# write out csv file and save a copy of data to another variable
px_data.to_csv("../dataset/px_data_id.csv")
px_data_save = px_data
```

```python
def try_pull_value(response, value: str, index: int, column_name) -> None:
    try:
        if value == "location":
            px_data.loc[index, column_name] = response[value]['display_address'][0]
        else:
            px_data.loc[index, column_name] = response[value]
    except:
        print(f"Value: {value} was not found in response at index: store {row[0]}")
```

```python
# use yelp id to find out detail of specific location
endpoint_search = "https://api.yelp.com/v3/businesses"
counter = 0
for index, row in px_data.iterrows():
    unique_id = row[10]
    
    response = requests.get(f"{endpoint_search}/{unique_id}", headers = header).json()
    
    if response.get("error",0):
        counter = counter+1
        print(f"---------Store {row[0]} invalid response. Total {counter} stores so far. ")

    else:
        try_pull_value(response=response, value="rating", index=index, column_name="yelp_rating")
        try_pull_value(response=response, value="price", index=index, column_name="price_tag")
        try_pull_value(response=response, value="review_count", index=index, column_name="review_count")
        try_pull_value(response=response, value="location", index=index, column_name="yelp_address")

```

```python
# pull yelp api again for empty cells
for index, row in px_data.iterrows():
    
    if pd.isnull(row[11]):
        unique_id = row[10]

        response = requests.get(f"{endpoint_search}/{unique_id}", headers = header).json()

        if response.get("error",0):
            counter = counter+1
            print(f"---------Store {row[0]} invalid response. Total {counter} stores so far. ")

        else:
            try_pull_value(response=response, value="rating", index=index, column_name="yelp_rating")
            try_pull_value(response=response, value="price", index=index, column_name="price_tag")
            try_pull_value(response=response, value="review_count", index=index, column_name="review_count")
            try_pull_value(response=response, value="location", index=index, column_name="yelp_address")

    
    time.sleep(.25)
```
